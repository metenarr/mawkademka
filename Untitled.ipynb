1.4 Формирование дополнительных атрибутов

Сформируем отдельный атрибут, в которых будет содержаться анализ распространения вируса с помощью коэффициента распространения инфекции.

```python
date = pd.to_datetime(date)
date
```

```python
df['previous_4_days'] = df['new_cases'].rolling(window=4).sum()
df['last_4_days'] = df['new_cases'].rolling(window=4).sum().shift(-3)
df.isnull().sum()/len(df) * 100
```

```python
#Заполняем пропуски
irretrievable_cols = list()
for column in df.columns:
    if df[column].isnull().sum() / len(df) * 100 > 0:
        irretrievable_cols = column
        mean_rpm = df[irretrievable_cols].mean()
        df[irretrievable_cols].fillna(mean_rpm, inplace=True)
```

```python
df['Rt'] = df['last_4_days']  / df['previous_4_days']
```

```python
df.isnull().sum() / len(df) * 100
```

```python
#Заполняем пропуски
irretrievable_cols = list()
for column in df.columns:
    if df[column].isnull().sum() / len(df) * 100 > 0:
        irretrievable_cols = column
        mean_rpm = df[irretrievable_cols].mean()
        df[irretrievable_cols].fillna(mean_rpm, inplace=True)
```

```python
#Создаем уровень риска в стране по новому критерию Rt
df['risk_level'] = np.where((df['Rt'] < 1),
                            'green',
                            np.where((df['Rt'] >= 1) &
                                     (df['Rt'] < 1.5),
                                     'yellow',
                                     'red'))
```

```python
risk_level = {'green': 0, 'yellow': 1, 'red': 2}
df['risk_level'] = df['risk_level'].map(risk_level)
df
```

```python
value = df['Rt'].quantile(0.98)
df = df.replace(np.inf, value)
value = df['Rt'].quantile(0.00)
df = df.replace(np.nan, value)
df
```

2.1 Разбиение набора данных

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns = 'risk_level'), df['risk_level'], test_size = 0.33, stratify = df['risk_level'])
```

```python
df['risk_level'].hist(bins=20, color='black');
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100, max_depth=5, n_jobs=-1)
rfc.fit(X_train,y_train)
```

```python
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
print(precision_score(y_test, rfc.predict(X_test), average='macro'))
print(recall_score(y_test, rfc.predict(X_test), average='macro'))
```

```python
featureImportance = pd.DataFrame({"feature": X_train.columns,
                                  "importance": rfc.feature_importances_})

featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(["importance"], ascending=False, inplace=True)
featureImportance["importance"].plot(kind='bar');
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3, n_jobs=-1)
knn.fit(X_train,y_train)
```

```python
print(precision_score(y_test, knn.predict(X_test), average='macro'))
print(recall_score(y_test, knn.predict(X_test), average='macro'))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth= 5, random_state=2023)
dtc.fit(X_train, y_train)
```

```python
print(precision_score(y_test, dtc.predict(X_test), average='macro'))
print(recall_score(y_test, dtc.predict(X_test), average='macro'))
```

```python
featureImportance = pd.DataFrame({"feature": X_train.columns,
                                  "importance": dtc.feature_importances_})

featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(["importance"], ascending=False, inplace=True)
featureImportance["importance"].plot(kind='bar');
```
